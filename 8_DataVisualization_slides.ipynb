# Matplotlib: Subplots

## Multiple Plots Side by Side

````python
import matplotlib.pyplot as plt
import numpy as np

# Random data
x          = np.linspace(0, 100, 50)
y1         = np.random.randint(0, 101, size=50)
y2         = np.random.uniform(0, 100, size=50)
categories = ['A', 'B', 'C', 'D', 'E']
values     = np.random.randint(0, 101, size=5)
data       = np.random.normal(loc=50, scale=15, size=500)

# 2 rows, 2 columns
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# First plot (top left)
axes[0, 0].plot(x, y1)
axes[0, 0].set_title('Plot 1 - line')

# Second plot (top right)
axes[0, 1].scatter(x, y2)
axes[0, 1].set_title('Plot 2 - scatter')

# Third plot (bottom left)
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Plot 3 - bar')

# Fourth plot (bottom right)
axes[1, 1].hist(data, bins=20)
axes[1, 1].set_title('Plot 4 - histogram')

plt.tight_layout()
plt.show()
````

# 2. Seaborn: Statistical Plots

## Elegant & Simple

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Random DataFrame df
np.random.seed(42)
n = 200
df = pd.DataFrame({
    'age':      np.random.randint(18, 80, size=n),
    'category': np.random.choice(['A', 'B', 'C'], size=n),
    'value':    np.random.normal(loc=50, scale=15, size=n),
    'score':    np.random.uniform(0, 100, size=n),
    'income':   np.random.normal(loc=3000, scale=800, size=n)
})

# Set style
sns.set_style('whitegrid')

# Histogram with KDE
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data=df, x='age', kde=True, ax=ax)
plt.tight_layout()
plt.show()

# Boxplot by category
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df, x='category', y='value', ax=ax)
plt.tight_layout()
plt.show()

# Violin plot: combi --> Boxplot + KDE
fig, ax = plt.subplots(figsize=(10, 6))
sns.violinplot(data=df, x='category', y='value', ax=ax)
plt.tight_layout()
plt.show()

# Correlation heatmap (nur numerische Spalten)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', ax=ax)
plt.tight_layout()
plt.show()
````

# 3. Plotly: Interactive Visualizations

## Zoom, Hover & More

```python
import plotly.express as px

# Interactive line plot
fig = px.line(df, x='date', y='value', title='Time Series')
fig.show()

# Interactive scatter with color
fig = px.scatter(df, x='x', y='y', color='category',
                 hover_data=['info'])
fig.show()

# Interactive boxplot
fig = px.box(df, x='category', y='value')
fig.show()
```

**Advantage:** Explore data interactively!

````python
import plotly.express as px
import pandas as pd
import numpy as np

# Random DataFrame erstellen
np.random.seed(42)
n = 100
df = pd.DataFrame({
    'date':     pd.date_range(start='2024-01-01', periods=n, freq='D'),
    'value':    np.random.randint(0, 101, size=n),
    'x':        np.random.uniform(0, 100, size=n),
    'y':        np.random.uniform(0, 100, size=n),
    'category': np.random.choice(['A', 'B', 'C'], size=n),
    'info':     [f'Punkt {i}' for i in range(n)]
})



# Interactive line plot
fig = px.line(df, x='date', y='value', title='Time Series')
fig.show()

# Interactive scatter with color
fig = px.scatter(df, x='x', y='y', color='category',
                 hover_data=['info'])
fig.show()

# Interactive boxplot
fig = px.box(df, x='category', y='value')
fig.show()
````

## Don'ts
✗ Too many colors  
✗ 3D charts (hard to read)  
✗ Too much information in one plot  
✗ Misleading axes (not starting at 0)
